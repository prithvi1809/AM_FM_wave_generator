```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
#Taking Inputs from User (input functions can be only trigonometry functions.)
mod_func=input("Enter the modulating function in terms of t: ") # eg-(sin(2*math.pi*t))
mod_index=input("Enter the modulating index: ")
carrier_freq=input("Enter the carrier frequency: ")
```

```python
#Return array of amplitudes of modulating wave .
def get_modulating_amplitude(time):
    amplitude=[]
    for t in time:
        amplitude.append(eval("np."+mod_func))
    return amplitude
```

```python
#Plotting modulating function\
def plot_func(amplitude,time,title):
    plt.figure(figsize=(7,3))
    horiz_line_data=np.array([0 for i in range(len(time))])
    plt.plot(time,amplitude)
    plt.plot(time,horiz_line_data,'k')
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.title(title)
    plt.show()
```
